# Hierarchical Clustering

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df= pd.read_csv('Mall_Customers.csv')
X=df.iloc[:,[3,4]]
```

```python
import scipy.cluster.hierarchy as sch
dendrogram=sch.dendrogram(sch.linkage(X,method='ward'))
plt.title('Dendrogram for HC')
plt.xlabel('Customers')
plt.ylabel('Euclidean Distance')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=5,metric='euclidean',linkage='ward')
y_hc = hc.fit_predict(X)
```

```python
print(X)
```

```python
plt.scatter(X.iloc[y_hc == 0,0],X.iloc[y_hc == 0,1],s=100, c='blue',label='Cluster1')
plt.scatter(X.iloc[y_hc == 1,0],X.iloc[y_hc == 1,1],s=100, c='red',label='Cluster2')
plt.scatter(X.iloc[y_hc == 2,0],X.iloc[y_hc == 2,1],s=100, c='yellow',label='Cluster3')
plt.scatter(X.iloc[y_hc == 3,0],X.iloc[y_hc == 3,1],s=100, c='green',label='Cluster4')
plt.scatter(X.iloc[y_hc == 4,0],X.iloc[y_hc == 4,1],s=100, c='pink',label='Cluster5')
plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
